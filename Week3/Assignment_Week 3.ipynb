1. Lakukan handling outlier pilih 1 kolom saja yang di handling dengan IQR setelah mengecek distribusinya
2. Dan cek distribusinya kembali

```python
df = pd.read_csv('company.csv')
df.head()
```

```python
### Assignment 1.b Lakukan handling outlier pilih 1 kolom saja yang di handling dengan IQR setelah mengecek distribusinya
### Dan cek distribusinya kembali
sns.histplot(df['Revenue'], kde=True)
plt.title('Distribution Before Outlier Handling')
plt.show()

sns.boxplot(x=df['Revenue'])
plt.title('Boxplot Before Outlier Handling')
plt.show()
```

```python
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')
```

```python
Q1 = df['Revenue'].quantile(0.25)
Q3 = df['Revenue'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower:", lower_bound)
print("Upper:", upper_bound)
```

```python
df_clean = df[(df['Revenue'] >= lower_bound) & (df['Revenue'] <= upper_bound)]
```

```python
sns.histplot(df_clean['Revenue'], kde=True)
plt.title('Distribution After Outlier Handling')
plt.show()

sns.boxplot(x=df_clean['Revenue'])
plt.title('Boxplot After Outlier Handling')
plt.show()
```

## Assignment 3.a Label Encoder :
1. Pilih kolom-kolom StreamingMovies, StreamingTV, TechSupport, DeviceProtection, OnlineBackup, OnlineSecurity, MultipleLines
2. Ganti nilai yang tidak sesuai : No internet service menjadi No lalu validasi dengan value_counts
3. Kemudian lakukan label Encoder
4. Notes : gunakan Telco-Customer-Churn.csv

```python
#Tulis jawaban
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('TelcoCustomerChurn.csv')
df.head()
```

```python
cols = [
    'StreamingMovies',
    'StreamingTV',
    'TechSupport',
    'DeviceProtection',
    'OnlineBackup',
    'OnlineSecurity',
    'MultipleLines'
]

df_selected = df[cols]
df_selected.head()
```

```python
for col in cols:
    print(f"\n{col}")
    print(df_selected[col].value_counts())
```

```python
df_selected = df_selected.replace('No internet service', 'No')
```

```python
for col in cols:
    print(f"\n{col}")
    print(df_selected[col].value_counts())
```

```python
le = LabelEncoder()

for col in cols:
    df_selected[col] = le.fit_transform(df_selected[col])
```

```python
for col in cols:
    print(f"\n{col} mapping:")
    print(dict(zip(le.classes_, le.transform(le.classes_))))
```
